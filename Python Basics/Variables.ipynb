```python
## Understanding variable types
## Python is a dynamically typed language, which means you don't need to declare the type of a variable when you create it. The type is inferred based on the value assigned to the variable.

age=25 #int
hieght=5.9 #float
name="Charlie" #string
is_student=True #boolean

print(type(age))
```

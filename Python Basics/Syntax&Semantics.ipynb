## "Syntax & Semantics in Python"

### Syntax referes to the set of rules that defines the combinations of symbols that are considered to be correctly structured programs in a language. In simpler terms , syntax is about the correct arrangements of words and symbols in a code.

### Semantics refers tomthe meaning or interpretation of the symbols, characters and commands in a language. It is about what the code is supposed to do when it runs.

```python
# Basic Syntax Rules in python
## Case Sensitivity- Python is case sensitive. For example, the variable name 'myVar' and 'myvar' would be considered different variables.

name ="ABC"
Name = "xyz"

print(name)
print(Name)
```

```python
## Indentation- Python uses indentation to define blocks of code. Consistent indentation is crucial for the correct execution of the code.
if 5 > 2:
    print("Five is greater than two!")
```

```python
## Single line comment

print("Hello, World!")  # This is a comment
```

```python
## Line continuation using backslash "\" - 

total = 1 + 2 + 3 + \
        4 + 5 + 6
print(total)
```

```python
## Multiple Statements in a single line using semicolon ";"
x = 5; y = 10; print(x + y)
```

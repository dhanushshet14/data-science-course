# Intro to lists in Python
A list is a collection which is ordered and changeable. In Python, lists are written with square brackets.


```python
lst=[]
print(type(lst))
```

```python
name =["John","Doe","Smith",1,2,3,4,]
print(name)
```

```python
mixed_list = ["Alice", 42, 3.14, True, None]
print(mixed_list)
```

```python
## accessing list items
fruits = ["apple", "banana", "cherry", "date", "kiwi"] # all the indexes in python start from 0 i.e., 1,,2,3,4,5...
print(fruits[0])      # First item
print(fruits[2])      # Third item
print(fruits[-1])     # Last item
print(fruits[-3])     # Third from last item
```

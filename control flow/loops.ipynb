# LOOPS

```python
## for loop
for i in range(5):
    print("Iteration:", i)  
    
```

```python
## strings iteration
string = "Hello"
for char in string:
    print("Character:", char)
    
```

```python
## While loop
## defination: A while loop repeatedly executes a block of code as long as a specified condition is true.
count = 0
while count < 5:
    print("Count:", count)
    count += 1
    
```

```python
## contiue and break statements
# Using continue to skip even numbers
for i in range(10):
    if i % 2 == 0:
        continue
    print("Odd Number:", i)

# Using break to exit loop when number is greater than 5
for i in range(10):
    if i > 5:
        break
    print("Number:", i)
        
```

```python
## pass statement : A pass statement is a null operation; it is used when a statement is syntactically required but no action is needed.
for i in range(5):
    if i == 3:
        pass  # Placeholder for future code
    print("Number:", i)
```

```python
# nested loops
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
```

```python
## example: calaculate the sum of first n natural numbers using loops while and for
n = 10
# Using for loop
sum_for = 0
for i in range(1, n + 1):
    sum_for += i
print("Sum using for loop:", sum_for)
# Using while loop
sum_while = 0
count = 1
while count <= n:
    sum_while += count
    count += 1
print("Sum using while loop:", sum_while)
```

```python
## common error and best practices
# Avoid using assignment operator (=) instead of comparison operator (==)
# Use proper indentation to avoid IndentationError
# Use parentheses to clarify complex conditions
# Example of a common error
x = 10
if x = 10:  # This will raise a SyntaxError
    print("x is 10")
# Correct usage
if x == 10:
    print("x is 10")
# Proper indentation
if x > 5:
    print("x is greater than 5")
    print("This line is properly indented")
# Using parentheses for clarity
if (x > 5 and x < 15) or (x == 20):
    print("x is between 5 and 15 or x is 20")
# Nested conditional statements
x = 25
if x > 10:
    if x < 30:
        print("x is between 10 and 30")
    else:
        print("x is 30 or more")
```
